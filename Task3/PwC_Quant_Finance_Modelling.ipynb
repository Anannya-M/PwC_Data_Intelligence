```python
# To compute the historical repayment percentages
#Since, the Unnamed column contains dates

df = data.drop(['Unnamed: 0'], axis = 1)
```

```python
percentages = round(df.div(df['Origination Amount'], axis = 'index') * 100 ,2)
percentages.head()
```

```python
# Now the origination amount has become 100% so we need it to revert back to its original value

percentages['Origination Amount'] = data['Origination Amount']
```

```python
percentages.head()
```

```python
# Now, we need to bring the unnamed column back to the percentages dataframe
index = 0 #index of unnamed column in data dataframe
percentages.insert(loc = index,column = 'Unnamed: 0', value = data['Unnamed: 0'])
```

```python
percentages.head()
```

```python
# Now we need to rename the column of Unnamed to Credit date

percentages = percentages.rename(columns = {'Unnamed: 0':'Credit Date'})
```

```python
percentages.head()
```

```python
# Now we need to compute the total sum of percentages 
#So, for that we need to drop the columns credit Date & Originatiion Amount

percentages_sum = percentages.drop(columns = ['Credit Date','Origination Amount'], axis = 0)
percentages_sum.head()
```

```python
#Now we will compute the sum of percentages in a new column Percent_Sum
percentages_sum['Percent_sum'] = percentages_sum.sum(axis = 1)
percentages_sum.head()
```

```python
#Computing the remaining amount from the origination Amount

percentages_sum['Remaining_Amount'] = percentages['Origination Amount']*(100 - percentages_sum['Percent_sum'])
remaining_amount = round(percentages_sum['Remaining_Amount'].sum(axis = 0),2)
```

```python
print(f'So, as of 31 December 2020, the remaining CHF is {remaining_amount}')
```
